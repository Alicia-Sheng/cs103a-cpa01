Question1: Which type of airplane has the fastest speed?

Quetion2: What is the relationship between models and speed?

```python
import pandas as pd
```

```python
df = pd.read_csv('data/feds1.csv')
```

```python
df.describe()
```

```python
print(list(df.columns))
```

```python
print(list(df.index))
```

```python
table = pd.pivot_table(df, values='speed', index=['timestamp'], columns=['model'])
table
```

```python
ax = table.plot.line(figsize=(15,8), title='Speed of Different Airplane Models between 081715 to 123115')
ax.set_xlabel('timestamp')
ax.set_ylabel('speed')
```

```python
group1 = df.groupby('model').mean().sort_values('speed', ascending=False)
group1
```

```python
print('The top five models with the highest speeds are ', end='')
for i in range(4):
    print(group1.index[i], end=', ')
print(str(group1.index[4]) + '.')
```

```python
group2 = df.groupby('type_aircraft').mean().sort_values('speed', ascending=False)
group2
```

```python
print('The type of aircraft with the highest speed is ' + str(group2.index[0]) + '.')
```

### Discussion

The airplanes with different models are having peak speeds in different times.

There is a relationship between airplane speed and model, and airplane speed and type.
